```python
from tkinter import *
from PIL import ImageTk, Image
from tkinter import filedialog
from tkinter import ttk
from keras.utils import load_img,img_to_array
from keras.models import load_model
import matplotlib.pyplot as plt
import time
import numpy as np
import os
a=Tk()
a.geometry('500x500+540+260')
a.title('Nhận Diện Bệnh Lá Trà')
a.resizable(False,False)
canvas_a = Canvas(a, width=500, height=500)
ba = ImageTk.PhotoImage(Image.open('C:/Users/dell/Desktop/AI project/b1.png'))
canvas_a.create_image(0, 0, anchor=NW, image=ba)
canvas_a.pack()
def open_b():
    a.withdraw()
    b.deiconify()
arrow = PhotoImage(file=r'C:\Users\dell\Desktop\AI project\\b4.png')
nut_arrow=Button(a,image=arrow,bd=0,command=open_b)
nut_arrow.place(x=100,y=400)
b = Toplevel(a)
b.title('Nhận Diện Bệnh Lá Trà')
b.geometry('1200x700+130+60')
b.resizable(False, False)
b.withdraw()
canvas_b = Canvas(b, width=1200, height=700)
bb = ImageTk.PhotoImage(Image.open("C:/Users/dell/Desktop/AI project/b2.png"))
canvas_b.create_image(0, 0, anchor=NW,image=bb)
canvas_b.pack()
fb=Frame(b,width=260, height=260)
fb.place(x=470,y=40)
bfb= ImageTk.PhotoImage(Image.open('C:/Users/dell/Desktop/AI project/b3.png'))
label_fb=Label(fb,image=bfb)
label_fb.pack()
file=0
def chon_anh():
    global bfb,file,solution
    name=filedialog.askopenfilename(initialdir='C:/Users/dell/Desktop/AI project/Test',title='Select A File',filetype=(('jpeg files','*.jpg'),('All','*.*')))
    file=str(name)
    image = Image.open(file)
    resized = image.resize((180, 180), Image.ANTIALIAS)
    bfb=ImageTk.PhotoImage(resized)
    label_fb.configure(image=bfb)
    nut_phathien.place(x=700,y=150)
    giaiphap.configure(text='')
    giaiphap_label.configure(text='')
    bt_gp.place_forget()
nut_chonanh=ttk.Button(b,text='Chọn ảnh đi',command=lambda :chon_anh())
nut_chonanh.place(x=475,y=250)
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
model_h5=load_model('C:/Users/dell/Desktop/AI project/Tea.h5')
result=''
ketqua = 'ketqua1'
def detect():
    global file,result,solution,ketqua
    img = load_img(file, target_size=(150, 150))
    plt.imshow(img)
    img = img_to_array(img)
    img = img.astype('float32')
    img = img / 255
    img = np.expand_dims(img, axis=0)
    kq = model_h5.predict(img)
    class_name = ['Algal leaf', 'Anthracnose', 'Bird eye spot', 'Brown Blight', 'Gray Blight', 'Healthy', 'Red Leaf Spot','White Spot']
    kq_kiemtra = int(np.argmax(kq, axis=1))
    print("Tình trạng lá:", class_name[kq_kiemtra])
    ketqua = class_name[kq_kiemtra]
    print(ketqua)
    time.sleep(1)
    giaiphap.configure(text='Tình trạng lá: {}'.format(class_name[kq_kiemtra]))
    bt_gp.place_configure(x=150,y=200)
nut_phathien=ttk.Button(b,text='Phát hiện bệnh',command=detect)
nut_phathien.place_forget()
giaiphap=Label(b,text=result,font=('Helvetica bold', 16))
giaiphap.place(x=90,y=70)
def giaiphap1():
    if ketqua =='Algal leaf':
        giaiphap_label.configure(text='Bệnh Algal nên phun thuốc diệt nấm 2 tuần 1 lần nếu điều kiện ẩm ướt kéo dài.'\
'\nThường xuyên thu dọn lá rụng và cắt tỉa lá bệnh ra khỏi cây. ')
    elif ketqua == 'Anthracnose':
        giaiphap_label.configure(text='Loại bỏ và tiêu hủy bất kỳ bộ phận nào của cây bị nhiễm bệnh.' \
                                         '\nCó thể phun thuốc trừ bệnh copper fungicide nhưng nên cẩn thận vì đồng có thể tích tụ ở mức độ độc hại trong đất.  ')
    elif ketqua == 'Bird Eye Spot':
        giaiphap_label.configure(text='Sử dụng Streptomyces sannanensis để kiểm soát mầm bệnh gây bệnh trên lá.')
    elif ketqua == 'Brown Blight':
        giaiphap_label.configure(text='Loại bỏ tất cả các lá bị ảnh hưởng và đốt chúng hoặc bỏ vào thùng rác.'\
'\nChe phủ xung quanh gốc cây bằng rơm, dăm gỗ hoặc các loại mùn tự nhiên khác để ngăn các bào tử nấm trong đất văng lên cây. ')
    elif ketqua == 'Gray Blight':
        giaiphap_label.configure(text='Loại bỏ và tiêu hủy các lá bị nhiễm bệnh.' \
                                     '\nĐiều trị nó càng sớm càng tốt và theo lịch trình: Sử dụng thuốc diệt nấm hữu cơ hoặc thuốc diệt nấm hóa học theo hướng dẫn.')
    elif ketqua == 'Healthy':
        giaiphap_label.configure(text='Lá khỏe mạnh, không cần giải pháp, chỉ cần tưới nước đầy đủ.')
    elif ketqua == 'Red Leaf Spot':
        giaiphap_label.configure(text='Trường hợp bệnh là một vấn đề dai dẳng, hãy áp dụng thuốc diệt nấm hiệu quả cứ sau 14 đến 21 ngày.' \
                                     '\nMột khi các triệu chứng của bệnh đốm lá đỏ xuất hiện, nên sử dụng hỗn hợp thuốc diệt nấm tiếp xúc và toàn thân để kiểm soát bệnh.')
    elif ketqua == 'White Spot':
        giaiphap_label.configure(text='Có thể bón đồng lỏng (một dung dịch được chứng nhận hữu cơ) lên các cây để kiểm soát nấm đốm trắng.' \
                 '\nPhun toàn bộ cây ngay khi thấy bệnh. Chú ý phun đều mặt trên và mặt dưới của lá, 7-10 ngày phun lại 1 lần.')
giaiphap_label=Label(b,text='',font=('Times New Roman',15))
giaiphap_label.place(x=90, y=300)
bt_gp=Button(b,text='Xem giải pháp',command=giaiphap1)
bt_gp.place_forget()
a.mainloop()
```
