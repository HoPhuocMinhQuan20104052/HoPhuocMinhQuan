```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 
temperature = ctrl.Antecedent(np.arange(20.4,41.3,0.1),'temperature')
windspeed = ctrl.Antecedent(np.arange(0.7,3.2,0.1),'windspeed')
rainfall = ctrl.Consequent(np.arange(0.2,9.2,0.1),'rainfall')
temperature['very low']=fuzz.trimf(temperature.universe,[20.4,20.4,23])
temperature['low']=fuzz.trimf(temperature.universe,[20.4,23,27.3])
temperature['normal']=fuzz.trimf(temperature.universe,[23,27.3,30])
temperature['high']=fuzz.trimf(temperature.universe,[27.3,30,35])
temperature['very high']=fuzz.trimf(temperature.universe,[30,41.2,41.2])
windspeed['very low']=fuzz.trimf(windspeed.universe,[0.7,0.7,1])
windspeed['low']=fuzz.trimf(windspeed.universe,[0.7,1,1.5])
windspeed['normal']=fuzz.trimf(windspeed.universe,[1,1.5,2.5])
windspeed['high']=fuzz.trimf(windspeed.universe,[1.5,2.5,3])
windspeed['very high']=fuzz.trimf(windspeed.universe,[2.5,3.1,3.1])
rainfall['very low']=fuzz.trimf(rainfall.universe,[0.2,0.2,2])
rainfall['low']=fuzz.trimf(rainfall.universe,[0.2,2,3.5])
rainfall['normal']=fuzz.trimf(rainfall.universe,[2,3.5,5])
rainfall['high']=fuzz.trimf(rainfall.universe,[3.5,5,7])
rainfall['very high']=fuzz.trimf(rainfall.universe,[5,9.1,9.1])
rule1 = ctrl.Rule(temperature['very low'] & windspeed['very low'], rainfall['very low'])
rule2 = ctrl.Rule(temperature['very low'] & windspeed['low'], rainfall['very low'])
rule3 = ctrl.Rule(temperature['very low'] & windspeed['normal'], rainfall['low'])
rule4 = ctrl.Rule(temperature['very low'] & windspeed['high'], rainfall['normal'])
rule5 = ctrl.Rule(temperature['very low'] & windspeed['very high'], rainfall['high'])
rule6 = ctrl.Rule(temperature['low'] & windspeed['very low'], rainfall['very low'])
rule7 = ctrl.Rule(temperature['low'] & windspeed['low'], rainfall['very low'])
rule8 = ctrl.Rule(temperature['low'] & windspeed['normal'], rainfall['low'])
rule9 = ctrl.Rule(temperature['low'] & windspeed['high'], rainfall['normal'])
rule10 = ctrl.Rule(temperature['low'] & windspeed['very high'], rainfall['high'])
rule11 = ctrl.Rule(temperature['normal'] & windspeed['very low'], rainfall['very low'])
rule12 = ctrl.Rule(temperature['normal'] & windspeed['low'], rainfall['low'])
rule13 = ctrl.Rule(temperature['normal'] & windspeed['normal'], rainfall['normal'])
rule14 = ctrl.Rule(temperature['normal'] & windspeed['high'], rainfall['high'])
rule15 = ctrl.Rule(temperature['normal'] & windspeed['very high'], rainfall['high'])
rule16 = ctrl.Rule(temperature['high'] & windspeed['very low'], rainfall['low'])
rule17 = ctrl.Rule(temperature['high'] & windspeed['low'], rainfall['low'])
rule18 = ctrl.Rule(temperature['high'] & windspeed['normal'], rainfall['normal'])
rule19 = ctrl.Rule(temperature['high'] & windspeed['high'], rainfall['high'])
rule20 = ctrl.Rule(temperature['high'] & windspeed['very high'], rainfall['low'])
rule21 = ctrl.Rule(temperature['very high'] & windspeed['very low'], rainfall['low'])
rule22 = ctrl.Rule(temperature['very high'] & windspeed['low'], rainfall['normal'])
rule23 = ctrl.Rule(temperature['very high'] & windspeed['normal'], rainfall['high'])
rule24 = ctrl.Rule(temperature['very high'] & windspeed['high'], rainfall['very high'])
rule25 = ctrl.Rule(temperature['very high'] & windspeed['very high'], rainfall['very high'])
rain_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
rain=ctrl.ControlSystemSimulation(rain_ctrl)
rain.input['windspeed']= 2.7
rain.input['temperature']= 25
rain.compute()
print(rain.output['rainfall'])
rainfall.view(sim= rain)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('Salary_Data.csv')
x=data.drop(['Salary'], axis=True)
y=data['Salary']
data.head()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
sns.pairplot(data)
plt.plot(x_train,y_train,'r.',label='Train Data')
plt.xlabel('year experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test).round(1)
cal=pd.DataFrame(np.c_[y_test,y_pred],columns=['Original Salary','Predicted Salary'])
print('Coeffient',model.coef_)
print('Intercept',model.intercept_)
print('Accuracy',model.score(x_test,y_test))
```

```python
plt.scatter(x_train,y_train, color = 'red')
plt.plot(x_train,model.predict(x_train), color = 'blue')
plt.title('Salary vs Experience(Train set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```
